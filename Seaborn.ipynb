```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

**Loading Data Set**

```python
iris = sns.load_dataset('iris')
iris.head()
```

**HistPlot**

```python
sns.histplot(iris['sepal_length'],bins=10)
plt.title('Histogram')
plt.show()
```

**PairPlot**

```python
sns.pairplot(iris,hue = 'species')
plt.show()
```

**HeatMap**

```python
iris_numeric = iris[["sepal_length","sepal_width","petal_length","petal_width"]]
corr_matrix = iris_numeric.corr()
sns.heatmap(corr_matrix,annot=True,cmap='viridis')
plt.title('Heat Map')
plt.show()
```

**Violinplot**

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='sepal_length',data=iris,color='lightyellow')
plt.title('Violin Plot')
plt.show()
```

**SwarmPlot**

**Swarm Plot**

```python
plt.figure(figsize=(6,3))
sns.swarmplot(x='species',y='sepal_width',data=iris,color='red')
plt.title('Swarm Plot')
plt.show()
```

**Strip Plot**

```python
plt.figure(figsize=(7,4))
sns.stripplot(x='species',y='petal_length',data=iris,color='lightgreen')
plt.title('Strip Plot')
plt.show()
```

**Box Plot**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal_width', data=iris,color='yellow')
plt.title('Box Plot')
plt.show()
```

**BarPlot**

```python
plt.figure(figsize=(5, 3))
sns.barplot(x='species', y='sepal_width', data=iris,color='violet',width=0.4)
plt.title('Bar Graph')
plt.show()
```

**Count Plot**

```python
plt.figure(figsize=(5, 3))
sns.countplot(x='species', data=iris,color='pink',width=0.5)
plt.title('Count Plot')
plt.show()
```

**Scatter Plot**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris,)
plt.title('Scatter Diagram')
plt.show()
```

**Line Chart**

```python
# For line chart, we'll aggregate the data first by species
plt.figure(figsize=(7, 4))
mean_values = iris.groupby('species').mean().reset_index()
sns.lineplot(x='species', y='sepal_width', data=mean_values,color='black',marker='s')
plt.title('Line Graph')
plt.show()
```
